```python
t,τ = var("t,τ")

x2 = 19
x3 = 16
A = 2 * ln(x2)
Tb = 3 * ln(2 * x3) / 4
T = Tb / 3
τ = var("τ")

def s1(t):
    return A * heaviside(t) - A * heaviside(t-T)
    
def s2(t):
    return A * heaviside(t-T) - A * heaviside(t-2*T)
    
def s3(t):
    return A * heaviside(t-2*T) - A * heaviside(t-3*T)
```

```python
plot(s1(t),t,-1,4,figsize=(3,2))
```

```python
plot(s2(t),t,-1,4,figsize=(3,2))
```

```python
plot(s3(t),t,-1,4,figsize=(3,2))
```

```python
A_psi = 2 / sqrt(ln(2 * x3))

def psi1(t):
    return A_psi * heaviside(t) - A_psi * heaviside(t-T)
    
def psi2(t):
    return A_psi * heaviside(t-T) - A_psi * heaviside(t-2*T)
    
def psi3(t):
    return A_psi * heaviside(t-2*T) - A_psi * heaviside(t-3*T)
```

```python
plot(psi1(t),t,-1,4,figsize=(3,2))
```

```python
plot(psi2(t),t,-1,4,figsize=(3,2))
```

```python
plot(psi3(t),t,-1,4,figsize=(3,2))
```

```python
def h1(t):
    return psi1(Tb - t)
    
def h2(t):
    return psi2(Tb - t)
    
def h3(t):
    return psi3(Tb - t)
```

```python
plot(h1(t),t,-1,4,figsize=(3,3),color="green",axes_labels=["$t$","$h_1(t)$"])
```

```python
plot(h2(t),t,-1,4,figsize=(3,3),color="green",axes_labels=["$t$","$h_2(t)$"])
```

```python
plot(h3(t),t,-1,4,figsize=(3,3),color="green",axes_labels=["$t$","$h_3(t)$"])
```

```python
t,τ = var("t,τ")
# assume(t > T/2 and t < T)

def y1s(t):
    return integral(s1(τ) * h1(t - τ),τ,0,t)

def y2s(t):
    return integral(s1(τ) * h2(t - τ),τ,0,t)

def y3s(t):
    return integral(s1(τ) * h3(t - τ),τ,0,t)
```

```python
plot(y1s(t),t,-1,4,figsize=(3,3),color="brown",axes_labels=["$t$","$y_{1s}(t)$"])
```

```python
plot(y2s(t),t,-1,4,figsize=(3,3),color="brown",axes_labels=["$t$","$y_{2s}(t)$"])
```

```python
plot(y3s(t),t,-1,4,figsize=(3,3),color="brown",axes_labels=["$t$","$y_{3s}(t)$"])
```

```python
ln_value = ln(2*x3)

r = (y1s(ln_value), y2s(ln_value), y3s(ln_value))
r
```

```python
N_0 = var("N_0")
s_1 = var("s_1")
m = 2 * ln(x2)
σ = sqrt(N_0 / 2)

def gaussian_pdf(r, mean, sigma):
    return (1 / sqrt(2 * pi * sigma**2)) * exp((-(r-m)**2) / (2 * sigma**2))

r = [y1s(m), y2s(m), y3s(m)]

conditional_prob = [gaussian_pdf(r_i, m, σ) for r_i in r]

f_r1_s1 = conditional_prob[0]
f_r2_s1 = conditional_prob[1]
f_r3_s1 = conditional_prob[2]
f_r_s1 = f_r1_s1 * f_r2_s1 * f_r3_s1
```
